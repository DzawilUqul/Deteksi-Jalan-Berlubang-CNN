```python
import gradio as gr
import tensorflow as tf
from keras.preprocessing import image
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
print(os.listdir('.'))
```

```python
model_path = "cnn.h5"
model = tf.keras.models.load_model(model_path)

def predict_pothole(image_file):
    # Load and preprocess the input image
    print(image_file)
    img_resized = image_file.resize((64, 64))
    test_image = image.img_to_array(img_resized)
    test_image = np.expand_dims(test_image, axis=0)

    # Make prediction
    result = model.predict(test_image)

    # Get class indices
    class_indices = {'normal': 0, 'pothole': 1}

    # Interpret the result
    if result[0][0] == 1:
        prediction = 'Jalan Berlubang'
    else:
        prediction = 'Normal'

    return prediction

# Create a Gradio interface
iface = gr.Interface(
    fn=predict_pothole,
    inputs=gr.Image(type="pil", label="Upload Image"),
    outputs="text",
    live=True,
    title="Pendeteksi Jalan Berlubang by Dzawil Uqul"
)

# Launch the interface
iface.launch(share=True, debug=True)
```
